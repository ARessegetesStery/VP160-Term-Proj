```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
K = 30000

seed = np.random.random((K , N))
step = np.where(seed < 0.5, -1, 1)
result = step.sum(axis=1)
```

```python
plt.hist(result, 40, rwidth=0.8)
plt.xlabel("Sailors' positions")
plt.ylabel("Number of sailors")
plt.show()
```

```python
sigma = ((result ** 2).mean() - result.mean() ** 2) ** 0.5
sigma
```

For the case $N = 100 , K = 30000$, in our simulation the value of $\sigma \left( x_{N} \right)$ is given above, which is close to $\sqrt{N} = 10$.

```python
def get_sigma(N, K=30000):
    seed = np.random.random((K , N))
    step = np.where(seed < 0.5, -1, 1)
    result = step.sum(axis=1)
    return ((result ** 2).mean() - result.mean() ** 2) ** 0.5
```

```python
N_vals = np.arange(1, 101)
sigma_vals = list(map(get_sigma, N_vals))
```

```python
plt.scatter(N_vals, sigma_vals, s=15)
plt.plot(N_vals, N_vals ** 0.5, "r")
plt.xlabel("Number of steps (N)")
plt.ylabel("Dispersion (sigma)")
plt.show()
```

We've just gone through all simulations with $N$ ranging from 1 to 100, and $K$ fixed to 30000. The scatter plot (blue dots) above demonstrates the relation between $\sigma \left( x_{N} \right)$ and $N$. It's clear that all these points form a graph close to the square root function. Therefore, our simulated result agrees with theoretical predictions.

As $\sigma \left( x_{N} \right)$ characterizes the dispersion of sailors' positions, we can draw a conclusion that the greater the number of steps sailors take, the larger disperation of their final positions will be, up to statistical point of view. Although the disperation becomes slower as $N$ gets larger, the rate of growth becomes slower.
